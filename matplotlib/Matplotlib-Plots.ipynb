# Plotting with Matplotlib

```python
import numpy as np

from tifffile import imread

from matplotlib import pyplot as plt
```

## Data Reading

```python
img = imread('../data/blobs.tif')
```

```python
print("Shape:", img.shape)
```

## Display an Image

```python
# Create a new figure with (width, height) = (10, 10)
plt.figure(figsize=(10,10))
plt.imshow(img);
```

### Configure Display

```python
# Create a new figure with (width, height) = (10, 10)
plt.figure(figsize=(10,10))
# Change the colormap to gray
plt.imshow(img)
# Remove the axis labels
plt.axis('off');
# Remove whitespace padding with tight_layout
plt.tight_layout()
```

### Colormaps

Matplotlib comes with a variaty of [colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html).

```python
# Create a new figure with (width, height) = (10, 10)
plt.figure(figsize=(10,10))
# Replace the default colormap with 'gray'
plt.imshow(img, cmap='gray')
# Remove the axis labels
plt.axis('off');
# Remove whitespace padding with tight_layout
plt.tight_layout()
```

### Save a Figure

```python
# Create a new figure with (width, height) = (10, 10)
plt.figure(figsize=(10,10))
# Change the colormap to inverted gray
plt.imshow(img, cmap='gray_r')
# Remove the axis labels
plt.axis('off');
# Remove whitespace padding with tight_layout
plt.tight_layout()
# The saved image has a small white border
plt.savefig('blobs_gray_r.png', pad_inches=0.0)
```

### Save Figure without Borders

```python
# Create figure
fig = plt.figure(figsize=(10,10), frameon=False)
# Add new axes object which spans the whole figure
ax = plt.Axes(fig, [0., 0., 1., 1.])
fig.add_axes(ax)
# Turn axes labels off
ax.set_axis_off()
# Display the image
ax.imshow(img, cmap='gray_r')
fig.savefig('blobs_gray_r_noBorder.png', bbox_inches=0.0)
```

## Subplots

Subplots is the concept of combining multiple plots in one figure. 

### Simple Subplots

The command `plt.subplot(nrows, ncols, index)` geneartes a grid of `nrows` x `ncols` subplots. Each subplot can be acces via index. 

Indexing:

1 2 3

4 5 6

```python
# Create figure
fig = plt.figure(figsize=(10, 5), frameon=False)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(img, cmap='gray_r')
plt.axis('off')

# Tight layout
plt.tight_layout()

fig.savefig('2_blobs.png', bbox_inches=0.0)
```

### Advanced Subplots

Sometimes you want to have subplots of different sizes. Using `plt.Axes(fig, [left, bottom, width, height])` allows you to add a subplot (axes) to figure `fig`. 

```python
# Create figure
fig = plt.figure(figsize=(15, 5), frameon=False)
# Add two new axes
ax0 = plt.Axes(fig, [0., 0., 0.31, 1.])
ax1 = plt.Axes(fig, [.35, 0.04, 0.65, 0.92])
fig.add_axes(ax0)
fig.add_axes(ax1)

# Display the image
ax0.imshow(img, cmap='gray')
# Turn axes labels of for the first subplot
ax0.set_axis_off()
# Set subplot title
ax0.set_title('Image')

# Display the image histogram
ax1.hist(img.flatten(), bins=np.linspace(0, 255, 10));
# Set the x-ticks to the bin-borders
plt.xticks(np.linspace(0,255,10));
# Format the y-ticks to abbreviate 1000 with k
plt.yticks(ticks=range(0,30001, 5000), labels=[str(i) + 'k' for i in range(0,31,5)])
# Set subplot title
ax1.set_title('Histogram')
# Add x-axis label
ax1.set_xlabel('Bins')

# Save the figure
fig.savefig('blobs_hist.png', bbox_inches=0.0)
```
